```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
Crime_Statitics=pd.read_csv('SouthAfricaCrimeStats_v2[1].csv')
```

```python
Crime_Statitics
```

```python
Crime_Statistics_by_province_yearly = Crime_Statitics.groupby('Province')[['2005-2006', '2006-2007', '2007-2008', '2008-2009', '2009-2010', '2010-2011', '2011-2012', '2012-2013', '2013-2014', '2014-2015', '2015-2016']]
```

```python
Crime_Statistics_by_province
```

```python

Crime_Statistics_by_province.T.plot(kind='bar', figsize=(12, 8))

plt.title('Crime Statistics by Province (2005-2016)')
plt.xlabel('Year')
plt.ylabel('Total Crime')

plt.legend(title='Province', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
Total=Crime_Statistics_by_province.sum(axis=1)
```

```python
Total.plot(kind='bar', figsize=(10, 6), color='blue')
plt.title('Total Crime Statistics by Province (2005-2016)')
plt.xlabel('Province')
plt.ylabel('Total Crime (2005-2016)')
plt.show(
```

```python
PoliceStations=Crime_Statitics.groupby('Province')['Station'].count()
```

```python
PoliceStations.plot(kind='bar', figsize=(12, 8))
plt.title('Total number of police stations per province')
plt.xlabel('Province')
plt.ylabel('Total police stations')
plt.show()
```

```python
categories=Crime_Statitics.groupby('Category')[['2005-2006', '2006-2007', '2007-2008', '2008-2009', '2009-2010', '2010-2011', '2011-2012', '2012-2013', '2013-2014', '2014-2015', '2015-2016']].sum()
```

```python
categories
```

```python
Total_categories=categories.sum(axis=1)
Total_categories
```

```python
Total_categories.plot(kind='bar', figsize=(12, 8), color='skyblue')

plt.title('Total Crimes per Crime Category (2005-2016)')
plt.xlabel('Crime Category')
plt.ylabel('Total Crimes')


plt.tight_layout()
plt.show()
```
